```python
# Splitting the Data into Training and Testing Sets
# Converting the values into zero's and one's
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
df_train,df_test = train_test_split(data,test_size=0.1,random_state=100)
```

```python
# Rescaling the features 
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler= MinMaxScaler()
```

```python
df_train.head()
```

```python
num_vars = ['Transaction_date', 'House_Age', 'Distance_metro', 'stores', 'latitude','longitude', 'bedrooms', 'sqft', 'price']
df_train[num_vars]= scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

```python
df_train.describe()
```

```python
sns.heatmap(df_train.corr(), annot=True)
```

```python
# Dividing into X and Y sets for the model building 
```

```python
y_train = df_train.pop('price')
X_train = df_train
```

```python
# Building a linear model 
```

```python
import statsmodels.api as sm

X_train_lm = sm.add_constant(X_train[['House_Age','Distance_metro','stores','latitude']])
lr_4 = sm.OLS(y_train,X_train_lm).fit()
print(lr_4.summary())
```

```python
# Recursive feature elimination 
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_absolute_percentage_error
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
rfe = RFE(lm,step=10)
rfe = rfe.fit(X_train,y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col=X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
X_train_rfe = X_train[col]
```

```python
import statsmodels.api as sm
X_train_rfe= sm.add_constant(X_train_rfe)
```

```python
lm = sm.GLS(y_train,X_train_rfe).fit()
print(lm.summary())
```

```python
# With the help of variance inflation we derive the correlated columns that decides and plays a major factor in the price prediction
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X=X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif=vif.sort_values(by="VIF",ascending = False)
vif
```

```python
num_vars = ['Transaction_date', 'House_Age', 'Distance_metro', 'stores', 'latitude','longitude', 'bedrooms', 'sqft', 'price']
df_test[num_vars]= scaler.transform(df_test[num_vars])
```

```python
df_test.describe()
```

```python
# we drop the columns that are not required
# Next we start to predict by dividing the X and y test
```

```python
y_test=df_test.pop('price')
X_test=df_test

X_test_m4 = sm.add_constant(X_test)

X_test_m4 = X_test_m4.drop(["Transaction_date", "longitude", "bedrooms", "sqft"], axis=1)
```

```python
# 1st prediction is using linear Regression
```

```python
y_pred_m4 = lr_4.predict(X_test_m4)
```

```python
fig = plt.figure()
plt.scatter(y_test,y_pred_m4)
fig.suptitle('y_test vs y_pred')
plt.xlabel('y_test')
plt.ylabel('y_pred')
```

```python
print(mean_absolute_percentage_error(y_test, y_pred_m4))
```

```python
# 2nd is by using SVM
```

```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.metrics import mean_absolute_percentage_error

 
model_SVR = svm.SVR()
model_SVR.fit(X_train_rfe,y_train)
Y_pred = model_SVR.predict(X_test_m4)
 
print(mean_absolute_percentage_error(y_test, Y_pred))
```

```python
fig = plt.figure()
plt.scatter(y_test,Y_pred)
fig.suptitle('y_test vs y_pred')
plt.xlabel('y_test')
plt.ylabel('y_pred')
```

```python
# 3rd is by using RandomForestRegressor
```

```python
from sklearn.ensemble import RandomForestRegressor
 
model_RFR = RandomForestRegressor(n_estimators=10)
model_RFR.fit(X_train_rfe, y_train)
Y_predd = model_RFR.predict(X_test_m4)
 
mean_absolute_percentage_error(y_test, Y_predd)
```

```python
fig = plt.figure()
plt.scatter(y_test,Y_predd)
fig.suptitle('y_test vs y_pred')
plt.xlabel('y_test')
plt.ylabel('y_pred')
```
